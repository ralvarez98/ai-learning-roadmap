```python
# 游늵 Visualizaci칩n de Datos con Matplotlib y Seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci칩n visual
sns.set(style="darkgrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Cargar el DataFrame limpio
df = pd.read_csv('../../data/ventas_limpio.csv')

# Mostrar las primeras filas
print("游늯 Vista previa del DataFrame:")
display(df)

# 游늷 Gr치fico 1: Barras de ingreso por categor칤a
ingresos_por_categoria = df.groupby('categoria')['ingreso'].sum().sort_values()

ingresos_por_categoria.plot(kind='barh', title='Ingresos Totales por Categor칤a')
plt.xlabel('Ingreso total ($)')
plt.ylabel('Categor칤a')
plt.tight_layout()
plt.show()

# 游늷 Gr치fico 2: Diagrama de dispersi칩n (precio vs cantidad)
sns.scatterplot(data=df, x='precio', y='cantidad', hue='categoria')
plt.title('Relaci칩n Precio vs Cantidad por Categor칤a')
plt.show()

# 游늷 Gr치fico 3: Histograma del precio
sns.histplot(data=df, x='precio', kde=True, bins=10)
plt.title('Distribuci칩n de Precios')
plt.show()

```

```python
# Cargar el DataFrame limpio
df = pd.read_csv('../../data/ventas_limpio.csv')

# Estad칤sticas generales del DataFrame
print("游늵 Estad칤sticas generales:")
print(df.describe())
# Agrupar por categor칤a
agrupado = df.groupby('categoria')

# Calcular media de precio y cantidad por categor칤a
print("游늵 Media por categor칤a:")
print(agrupado[['precio', 'cantidad']].mean())

# Promedio de ingreso por categor칤a
ingreso_categoria = df.groupby('categoria')['ingreso'].mean()

# Gr치fico de barras
ingreso_categoria.plot(kind='bar', title='Ingreso Promedio por Categor칤a')
plt.xlabel('Categor칤a')
plt.ylabel('Ingreso promedio')
plt.grid(True)
plt.tight_layout()
plt.show()
```
