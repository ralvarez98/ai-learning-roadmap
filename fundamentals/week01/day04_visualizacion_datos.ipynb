```python
# 📊 Visualización de Datos con Matplotlib y Seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración visual
sns.set(style="darkgrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Cargar el DataFrame limpio
df = pd.read_csv('../../data/ventas_limpio.csv')

# Mostrar las primeras filas
print("📄 Vista previa del DataFrame:")
display(df)

# 📌 Gráfico 1: Barras de ingreso por categoría
ingresos_por_categoria = df.groupby('categoria')['ingreso'].sum().sort_values()

ingresos_por_categoria.plot(kind='barh', title='Ingresos Totales por Categoría')
plt.xlabel('Ingreso total ($)')
plt.ylabel('Categoría')
plt.tight_layout()
plt.show()

# 📌 Gráfico 2: Diagrama de dispersión (precio vs cantidad)
sns.scatterplot(data=df, x='precio', y='cantidad', hue='categoria')
plt.title('Relación Precio vs Cantidad por Categoría')
plt.show()

# 📌 Gráfico 3: Histograma del precio
sns.histplot(data=df, x='precio', kde=True, bins=10)
plt.title('Distribución de Precios')
plt.show()

```

```python
# Cargar el DataFrame limpio
df = pd.read_csv('../../data/ventas_limpio.csv')

# Estadísticas generales del DataFrame
print("📊 Estadísticas generales:")
print(df.describe())
# Agrupar por categoría
agrupado = df.groupby('categoria')

# Calcular media de precio y cantidad por categoría
print("📊 Media por categoría:")
print(agrupado[['precio', 'cantidad']].mean())

# Promedio de ingreso por categoría
ingreso_categoria = df.groupby('categoria')['ingreso'].mean()

# Gráfico de barras
ingreso_categoria.plot(kind='bar', title='Ingreso Promedio por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Ingreso promedio')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ✅ Asegúrate de haber cargado tu DataFrame
# Por ejemplo:
df = pd.read_csv('../../data/ventas_limpio.csv')

# ✅ Si no lo habías hecho aún, crea la columna 'ingreso'
df['ingreso'] = df['precio'] * df['cantidad']

# ✅ Agrupamos por producto y sumamos el ingreso total por cada uno
ingreso_por_producto = df.groupby('producto')['ingreso'].sum().sort_values(ascending=False)

# ✅ Graficamos el resultado
ingreso_por_producto.plot(kind='bar', title='Ingreso total por producto', figsize=(10, 6))
plt.xlabel('Producto')
plt.ylabel('Ingreso Total')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
