üß™ Mini pr√°ctica D√≠a 7: Predicci√≥n del Salario seg√∫n la Experiencia
Contexto:
Imagina que est√°s colaborando con el √°rea de Recursos Humanos de una empresa tecnol√≥gica. Te piden analizar los datos de empleados y desarrollar un modelo que prediga el salario mensual con base en los a√±os de experiencia de un candidato.

```python
# Paso 1: Importaci√≥n de librer√≠as
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

sns.set(style='whitegrid')

# Paso 2: Crear un DataFrame simulado (como si fueran datos de empleados)
np.random.seed(42)
data = {
    'a√±os_experiencia': np.random.randint(1, 11, 50),
    'salario_mensual': np.random.randint(15000, 50000, 50)
}
df = pd.DataFrame(data)
df.head()

# Paso 3: Normalizaci√≥n con MinMaxScaler
scaler = MinMaxScaler()
df[['a√±os_experiencia', 'salario_mensual']] = scaler.fit_transform(df[['a√±os_experiencia', 'salario_mensual']])
df.head()

# Paso 4: Visualizaci√≥n de correlaciones
plt.figure(figsize=(5, 4))
sns.heatmap(df.corr(), annot=True, cmap='Blues')
plt.title('Matriz de correlaci√≥n')
plt.show()

# Paso 5: Entrenamiento del modelo
x = df[['a√±os_experiencia']]
y = df['salario_mensual']
modelo = LinearRegression()
modelo.fit(x, y)

# Paso 6: Predicci√≥n y creaci√≥n de nueva columna
df['salario_predicho'] = modelo.predict(x)

# Paso 7: Evaluaci√≥n del modelo
r2 = r2_score(y, df['salario_predicho'])
print(f"üîé R¬≤ Score del modelo: {r2:.4f}")

# Paso 8: Visualizaci√≥n del modelo entrenado
plt.figure(figsize=(7, 4))
sns.scatterplot(x='a√±os_experiencia', y='salario_mensual', data=df, label='Real')
sns.lineplot(x='a√±os_experiencia', y='salario_predicho', data=df, color='red', label='Modelo Lineal')
plt.title('Regresi√≥n Lineal: Salario vs A√±os de Experiencia')
plt.xlabel('A√±os de Experiencia (normalizado)')
plt.ylabel('Salario Mensual (normalizado)')
plt.legend()
plt.show()
```
